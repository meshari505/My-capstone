```python
import numpy as np
import pandas as pd
df=pd.read_csv(r'C:\my-capstone\data\cleaned\ames_features.csv')
prices = df['SalePrice'].values

#حساب الانحراف المعياري والمتوسط
mean_val = np.sum(prices) / len(prices)
std_val = np.sqrt(np.sum((prices - mean_val)**2) / len(prices))

print(f"المتوسط الحسابي: {mean_val:.2f}")
print(f"الانحراف المعياري: {std_val:.2f}")

#حساب تشابه الcos
row1 = df.select_dtypes(include=[np.number]).iloc[0].values
row2 = df.select_dtypes(include=[np.number]).iloc[1].values
cos_sim = np.dot(row1, row2) / (np.linalg.norm(row1) * np.linalg.norm(row2))
print(f"تشابه جيب التمام: {cos_sim:.4f}")
```
