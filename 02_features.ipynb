```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('data/cleaned/ames_cleaned.csv')

# 1. One-hot encoding (تعديل الاسم إلى 'House Style')
df = pd.get_dummies(df, columns=['Neighborhood', 'House Style'], drop_first=True)

# استخدمت SCALING for Lot area
scaler = StandardScaler()
df[['Lot Area']] = scaler.fit_transform(df[['Lot Area']])

#انشاء عمود جديد يحمل عمر المنزل
df['House_Age_At_Sale'] = df['Yr Sold'] - df['Year Built']

# تفاعل بين الجودة الكلية والمساحة (إذا كان اسم المساحة Gr Liv Area)
if 'Gr Liv Area' in df.columns:
    df['Quality_Area_Interaction'] = df['Overall Qual'] * df['Gr Liv Area']

# 4. Binning لعمر المنزل
df['Age_Group'] = pd.cut(df['House_Age_At_Sale'], bins=[-1, 10, 30, 200], labels=['New', 'Moderate', 'Old'])
#حفظ التغييرات الي سويتها بهذا الملف 
df.to_csv('data/cleaned/ames_features.csv', index=False)
```
