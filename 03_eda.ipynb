```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df=pd.read_csv(r'C:\my-capstone\data\cleaned\ames_features.csv')
#علاقة الجودة بالسعر
plt.figure(figsize=(10,6))
sns.boxplot(x='Overall Qual', y='SalePrice', data=df)
plt.title('Sale Price vs Overall Quality')
plt.show()

#تشتت اسعار المنازل
plt.figure(figsize=(8,5))
sns.histplot(df['SalePrice'], kde=True)
plt.title('Distribution of Sale Prices')
plt.show()

# تاثير عمر المنزل على سعره
plt.figure(figsize=(8,5))
sns.scatterplot(x='House_Age_At_Sale', y='SalePrice', data=df)
plt.title('House Age vs Sale Price')
plt.show()
```
