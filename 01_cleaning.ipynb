```python
import pandas as pd
import numpy as np

df = pd.read_csv(r'C:\my-capstone\data\raw\AmesHousing.csv')

def clean_data(raw_df):
    df_clean = raw_df.copy()
    
    # اشيل الاعمدة المتكررة
    df_clean = df_clean.drop_duplicates()
    
#معالجة البيانات المفقودة
    numeric_cols = df_clean.select_dtypes(include=[np.number]).columns
    df_clean[numeric_cols] = df_clean[numeric_cols].fillna(df_clean[numeric_cols].median())
    
#معالجة القيم المتطرفة
    upper_limit = df_clean['SalePrice'].quantile(0.99)
    df_clean['SalePrice'] = np.where(df_clean['SalePrice'] > upper_limit, upper_limit, df_clean['SalePrice'])
    
    return df_clean

df_cleaned = clean_data(df)

# التحقق
assert df_cleaned['SalePrice'].isnull().sum() == 0
assert (df_cleaned['SalePrice'] > 0).all()

# حفظ التغيرات والملف
df_cleaned.to_csv('data/cleaned/ames_cleaned.csv', index=False)
```
